```python
df= pd.read_csv("Null_clean.csv")
df
```

```python
valores_nulos= df.isnull().sum()
valores_nulos
```

```python
cuantitativas = df.select_dtypes(include=['number'])
cuantitativas
```

```python
cualitativas = df.select_dtypes(exclude=['number'])
cualitativas
```

```python
#Diagrama de caja o bigote de cada columna del dataframe para revisión de valores Atipicos
fig = plt.figure(figsize =(15, 8))
cuantitativas.plot(kind='box', vert=False)
plt.title("Valores Atípicos del Dataframe")
plt.show() #dibujamos el diagrama
```

```python
y=cuantitativas
Limite_Superior= y.mean() + 3*y.std()
Limite_Inferior= y.mean() - 3*y.std()
print("Limite superior permitido", Limite_Superior)
print("Limite inferior permitido", Limite_Inferior)
```

```python
Dato_or= cuantitativas[(y<=Limite_Superior)&(y>=Limite_Inferior)]
Dato_or
```

```python
data_clean=Dato_or.copy()
data_clean=data_clean.fillna(round(Dato_or.mean(),1))
data_clean
```

```python
#Diagrama de caja o bigote de cada columna del dataframe para revisión de valores Atipicos
fig = plt.figure(figsize =(15, 8))
data_clean.plot(kind='box', vert=False)
plt.title("Valores Atípicos del Dataframe")
plt.show() #dibujamos el diagrama
```

```python
Da_fi = pd.concat([cualitativas, data_clean], axis=1)
Da_fi
```

```python
Da_fi.to_csv("exact_data_reto.csv")
```
